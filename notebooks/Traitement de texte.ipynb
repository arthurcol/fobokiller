# Fonction traitement de texte

```python
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#--------------------------#
#first time using NLTK
# import nltk
#nltk.download('stopwords')
#nltk.download('punkt')
#nltk.download('wordnet')
#--------------------------#a
def clear_data_text(text, language):
    #remove punctuation and add space
    for punctuation in string.punctuation:
        text = text.replace(punctuation, ' ')
    # delete all non ascii characters and keep space
    text = ''.join(word for word in text if word.isalpha() or word ==' ')
    #pass in lowercase
    text = text.lower()
    #remove stop words
    stop_words = set(stopwords.words(language))
    #text to list of word
    word_tokens = word_tokenize(text)
    #list to string
    text = [w for w in word_tokens if not w in stop_words]
    return text
```
