```python
from sklearn.datasets import make_blobs
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X,y,centers = make_blobs(n_samples=10000,n_features=2,centers=7,cluster_std=2,shuffle=True,center_box=(-20,20),return_centers=True)
```

```python
colors=plt.cm.rainbow(np.linspace(0, 1, 7))
```

```python
plt.figure(figsize=(15,10))
plt.scatter(X[:,0],X[:,1],c=pd.Series(y).map({i:c for i,c in enumerate(colors)}),alpha=.5,s=4)
labels=['italian','atmosphere','french','expensive','luxe','fast-food','cafe']
for l,c in zip(labels,centers):
    plt.annotate(l,c,fontsize=14, fontweight='bold')
```
